```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import math
import warnings
warnings.filterwarnings('ignore')
```

```python
#Reading data files
file1 = pd.read_csv('lab-customer-analysis-round-1/files_for_lab/csv_files/file1.csv')
file2 = pd.read_csv('lab-customer-analysis-round-1/files_for_lab/csv_files/file2.csv')
file3 = pd.read_csv('lab-customer-analysis-round-1/files_for_lab/csv_files/file3.csv')
```

```python
#Storing and rearraging columns of each file

df1 = pd.DataFrame(data = file1)
df2 = pd.DataFrame(data = file2)
df3 = pd.DataFrame(data = file2)
#print(df1.shape,df1.shape,df2.shape)
#df1.info()
#df.info()
#df2.info()
#file1.header()
col1 = (df1.columns)
col1 = [val.lower().replace(' ','_') for val in col1]
df1.columns = col1
#print(df1.columns)
col2 = (df2.columns)
col2 = [val.lower().replace(' ','_') for val in col2]
df2.columns = col2
#print(df2.columns)
col3 = (df3.columns)
col3 = [val.lower().replace(' ','_') for val in col3]
df3.columns = col3
df2.columns = df1.columns
df3.columns = df1.columns
print(df3.columns == df1.columns)
#df1.head(1)
```

```python
#Concatanating
df = pd.DataFrame(columns= df1.columns)
df = pd.concat([df,df1,df2,df3], axis=0)
df.shape

#X_num = df.select_dtypes(include = np.number)
#X_Cat = df.select_dtypes(include = np.object)

#deleting the asked columns
del df['education']
del df['number_of_open_complaints']

```

```python
#Duplicate dropping
df = df.drop_duplicates()
df.shape
```

```python
# Data with income <= 0
df[df['income']<= 0]
```

```python
#cleaning customer_life_time_value data
df['customer_lifetime_value'] = df['customer_lifetime_value']*100
df['customer_lifetime_value'] =  pd.to_numeric(df['customer_lifetime_value'], errors='coerce')

```
