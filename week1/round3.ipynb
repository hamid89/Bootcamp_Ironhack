```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
import math
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
#Reading file
data = pd.read_csv('marketing_customer_analysis.csv')
#data.head(0)
```

```python
#data cleaning function from previous exercise
def my_func(df):
    df.columns=[e.lower().replace(' ', '_') for e in df.columns]
    df=df.dropna()
    df['effective_to_date']=pd.to_datetime(df['effective_to_date'], errors='coerce')
    return df
```

```python
data = my_func(data) 
```

```python
data.info(memory_usage='deep')
```

```python
data.describe()

data.describe(include='all')
```

```python
#Response plot
sns.countplot('response', data=data)
plt.ylabel('Total number of Response')
plt.show()
```

```python
#reponse rate by sales channel plot
plt.figure(figsize=(8,4))
sns.countplot('response', hue='sales_channel', data=data)
plt.ylabel('Response by Sales Channel')
plt.show()
```

```python
#plot of the response rate by the total claim amount.
plt.figure(figsize=(6,6))
sns.boxplot(y='total_claim_amount' , x='response', data=data)
plt.ylabel('Response by Total Claim Amount')
plt.show()
```

```python
#plot of the response rate by income.
plt.figure(figsize=(6,6))
sns.boxplot(y='income' , x='response', data=data)
plt.ylabel('Response by Inncome')
plt.show()
```
