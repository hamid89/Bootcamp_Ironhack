```python
engine.execute("USE sakila")
query = '''select COUNT(return_date_may), customer_id_may
from sakila.rentals_may
group by customer_id_may
order by customer_id_may asc'''
data1_may = pd.read_sql_query(query, engine)
data1_may.head(10)
```

```python
engine.execute("USE sakila")
query = '''select COUNT(return_date_june), customer_id_june
from sakila.rentals_june
group by customer_id_june
order by customer_id_june asc'''
data1_june = pd.read_sql_query(query, engine)

data1_june.head(10)
```

```python
def my_function(x1,x2):
    ls = []
    ls1 = []
    for val1,val2 in zip(x1,x2):
        ls.append(x1[val1] - x2[val2])
        if (x1[val1] - x2[val2]) < 0:
            ls1.append('more in June')
        elif (x1[val1] - x2[val2]) > 0:
            ls1.append('less in june')
        else:
                ls1.append('same in June and May')
    data_collective1 = pd.DataFrame(ls1, columns =['business_status'])
    data_collective2 = pd.DataFrame(ls, columns =['business_difference'])
    #data_collective = pd.DataFrame(columns = ['customer_id','business_status','business_difference'])
    data_collective = pd.concat([data1_june['customer_id_june'],data_collective1['business_status'],data_collective2['business_difference']],axis = 1)#,columns = ['customer_id','business_status','business_difference'])
    
    
    return data_collective
```

```python
f = my_function(data1_may['COUNT(return_date_may)'],data1_june['COUNT(return_date_june)'])
f.head(20)
```
